```python
import pandas as pd
import sqlite3 as db
import numpy as np

import datetime
from datetime import datetime, timedelta, date
```

```python
def rental_availability():
    '''
    Joins the LuxuryAPI, Regular API, Kaggle CSV, and the Webscraping events into one query. It then returns rental availability
    and an event schedule based on the date provided. 

    Inputs:

    LuxuryAPI table
    AllCarsAPI table
    API table
    Kaggle table
    WebScrape table
    LargeEvents table    

    Returns:

    A single query with rental availability
        
    '''

    # YOUR CODE HERE
    conn = db.connect('rentaldb')
    c = conn.cursor()
    query = f'''
        SELECT
            AllCarsAPI.Date_Pickup, AllCarsAPI.Vehicle_Name, Kaggle.Vehicle_Name, COUNT(DISTINCT AllCarsAPI.Vehicle_Name) AS Vehicle_Stock, Kaggle.daily_rate, AllCarsAPI.Price, COUNT(DISTINCT Webscrape.Event) AS Number_Of_Events, COUNT(DISTINCT LargeEvents.Dates) AS Number_OF_Large_Events
            
        FROM 
            AllCarsAPI
            FULL JOIN Kaggle ON AllCarsAPI.Vehicle_Name = Kaggle.Vehicle_Name
            FULL JOIN Webscrape ON AllCarsAPI.Date_Pickup = Webscrape.Dates
            FULL JOIN LargeEvents ON AllCarsAPI.Date_Pickup = LargeEvents.Dates
        GROUP BY 
            AllCarsAPI.Vehicle_Name 
        ORDER BY
            AllCarsAPI.Date_Pickup DESC
            '''
            
    
    #c.execute(query)
    #availability = c.fetchall()
    availabilitydf = pd.read_sql_query(query, conn)
    return availabilitydf
```

```python
availabilitydf = rental_availability()
display(availabilitydf.head(50))
```

```python
def rental_availability_API():
    # YOUR CODE HERE
    conn = db.connect('rentaldb')
    c = conn.cursor()
    query = f'''
        SELECT
            DISTINCT AllCarsAPI.Vehicle_Name, COUNT(AllCarsAPI.Vehicle_Name)
        FROM 
            AllCarsAPI
        GROUP BY 
            AllCarsAPI.Vehicle_Name
            '''
    availabilitydf = pd.read_sql_query(query, conn)
    return availabilitydf
```

```python
availabilitydf = rental_availability_API()
display(availabilitydf.head(50))
```

```python
def rental_availability_CSV():
    # YOUR CODE HERE
    conn = db.connect('rentaldb')
    c = conn.cursor()
    query = f'''
        SELECT
            DISTINCT Kaggle.Vehicle_Name, COUNT(Kaggle.Vehicle_Name)
        FROM 
            Kaggle
        GROUP BY 
            Kaggle.Vehicle_Name
            '''
    availabilitydf = pd.read_sql_query(query, conn)
    return availabilitydf
```
